````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
````

````python
def calculate_total_price(products):
    total_price = 0
    for product in products:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}:"))
                if price < 0:
                    raise ValueError("Invalid price, please enter a positive number")
                valid_price = True
            except ValueError as e:
                print(f"Error:{e}")
        total_price += price
    return total_price


def get_customer_orders(inventory):
    orders = {}
    valid_order = False
    while not valid_order:
        try:
            num_orders = int(input("Enter the number of orders:"))
            if num_orders < 0:
                raise ValueError(
                    "Invalid number of orders, please enter a positive number"
                )
            valid_order = True
        except ValueError as e:
            print(f"Error.{e}")

        for i in range(num_orders):
            valid_product = False
            while not valid_product:
                product_name = input("Enter the product name:")
                if product_name in inventory and inventory[product_name] > 0:
                    valid_product = True
                    orders[product_name] = orders.get(product_name, 0) + 1
                    inventory[product_name] -= 1
                else:
                    print(
                        "Invalid product name or product out of stock. Please try again."
                    )
    return orders

````

````python
# Example usage
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products)
print("Inventario inicializado:", inventory)

calculate_total_price(products)
print("Total price:", calculate_total_price(products))

orders = get_customer_orders(inventory)
print("Pedidos realizados:", orders)
````
